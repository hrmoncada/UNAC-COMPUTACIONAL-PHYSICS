## Step 3. Go through our HDP Hello World

Python code works here the same way that it does in a standard Jupyter notebook.

```
import os

import numpy as np
import pickle

import src.visualization.visualize_hdpmm as viz
from src.models.conjugate_pairs import CategoricalConjugatePair
from src.models.hdp_mixture_model import HdpMixtureModel as HDPMM
```

```
module_base_dir = '/content/drive/My Drive/module0colab'
```

```
# load data from our shared resource
data_file = f'{module_base_dir}/data/dataset.pkl'
with open(data_file, mode='rb') as f:
    R = pickle.load(f)
```

The data are a drawn according to an HDP.

```
data = R['data']
vocab_size = R['vocab_size']
n_groups = len(data)
```

Create the conjugate pair for categorical likelihood/Dirichlet prior.

```
cp = CategoricalConjugatePair(
    concentration=np.ones(vocab_size) / vocab_size)
```

Create a directory for model output.  It is **imperative** that this is NOT under the shared colab folder.  We create an output directory "module0output" under the user's My Drive folder.  That way different users do not end up writing to the same file.  Further, it will not conflict with output directories from other modules.  It will sort in alphabetical order to be nearby the `module0colab` folder.

```
output_dir = f'/content/drive/My Drive/module0output/models/hdp_demo'
os.makedirs(output_dir, exist_ok=True)
```

Fit an HDP topic model to the loaded dataset.

```
model = HDPMM(cp, fit_burn_in=100, fit_samples=200,
              history=f'{output_dir}/samples.h5')
labels = model.fit_predict(data)
n_atoms = model.atoms.shape[0]
```

Save a visualization of the local topic weights to your personal output directory.  Under your "My Drive" this file is created in `module0output/models/hdp_demo`.

```
# provide a save_dir to tell our visualization function to save the output
viz.plot_local_weights_image(model.local_weights,  save_dir=output_dir)
```

Produce other visualizations, but don't save them.

```
log_x = False
viz.plot_global_weights_image(model.global_weights)

viz.plot_topic_parameters(model.atoms, xlabel='word index')
viz.plot_topic_similarities(model.atoms)

viz.plot_local_word_likelihoods_image(
    model.local_weights, model.atoms)

viz.plot_local_word_likelihood_similarities_image(
    model.local_weights, model.atoms)

viz.plot_loglik(model.history, log_x=log_x)
viz.plot_num_topics_trace(model.history, log_x=log_x)
viz.plot_num_topics_hist(model.history)

viz.plot_concentration_hist(model.history)
viz.plot_concentration_trace(model.history, log_x=log_x)
viz.plot_base_measure_assignment_trace(model.history, log_x=log_x)
```
